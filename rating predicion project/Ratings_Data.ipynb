#### Importing required libraries: 

```python
import selenium
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from tqdm.notebook import tqdm_notebook
import pandas as pd
```

#### Scrapping reviews and ratings form the e-commerce website flipkart

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
time.sleep(3)

# Opening the homepage of Amazon.in
url = "https://www.flipkart.com/"
driver.get(url)

try:
    login_X_button = driver.find_element_by_xpath('//button[@class="_2KpZ6l _2doB4z"]')                      # Button to close login popup
    login_X_button.click()
except NoSuchElementException : 
    print("No Login page")

time.sleep(2)
# Asking the user to input the keywords he/she wants to search
user_inp = input('Enter the product you want to search : ')

search_bar = driver.find_element_by_xpath('//*[@id="container"]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')    # Finding the search bar using it's xpath
search_bar.clear()  # Clearing the search bar
search_bar.send_keys(user_inp)       # Inputing keyword to search
search_button = driver.find_element_by_xpath('//button[@class="L0Z3Pu"]')    # Finding the xpath of search button
search_button.click()        # Clicking the search button
```

```python
# Scraping url of all page
product_url = []
start = 0
end = 10

for page in range(start,end):
    url = driver.find_elements_by_xpath('//a[@class="s1Q9rs"]')
    for i in url:
        product_url.append(i.get_attribute('href'))
    nxt_button=driver.find_elements_by_xpath("//a[@class='_1LKTO3']")
    try:
        driver.get(nxt_button[1].get_attribute('href'))     #getting the link from the list for next page
    except:
        driver.get(nxt_button[0].get_attribute('href'))
    time.sleep(5)
    
# Creating empty list to store data
Reviews = []
Ratings = []

# Scraping data from each url
for i in product_url:
    driver.get(i)
    print("Scraping URL:", i)
    
    try:
        ratings_tag = driver.find_element_by_xpath('//div[@class="_3LWZlK _1BLPMq"]') #Extracting the ratings
        Ratings.append(ratings_tag.text)
    except NoSuchElementException:
        Ratings.append('No Rating')
    
    try:
        reviews_tag = driver.find_element_by_xpath('//div[@class="t-ZTKy"]') #Extracting the reviews
        Reviews.append(reviews_tag.text.replace('\n',''))
    except NoSuchElementException:
        Reviews.append('No Review')
```

#### Scrapping reviews and ratings of headphones form the e-commerce website myntra

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
time.sleep(3)

# Opening the homepage of Myntra.com
url = "https://www.myntra.com/"
driver.get(url)

time.sleep(2)
# Asking the user to input the keywords he/she wants to search
user_inp = input('Enter the product you want to search : ')

search_bar = driver.find_element_by_xpath('//input[@class="desktop-searchBar"]')    # Finding the search bar using it's xpath
search_bar.clear()  # Clearing the search bar
search_bar.send_keys(user_inp)       # Inputing keyword to search
search_button = driver.find_element_by_xpath('//a[@class="desktop-submit"]')    # Finding the xpath of search button
search_button.click()        # Clicking the search button
```

```python
# Scraping url of all page
product_url = []
start = 0
end = 8

for page in range(start,end):
    url = driver.find_elements_by_xpath('//li[@class="product-base"]//a')
    for i in url:
        product_url.append(i.get_attribute('href'))
    nxt_button=driver.find_elements_by_xpath('//li[@class="pagination-next"]//a')
    try:
        driver.get(nxt_button[1].get_attribute('href'))     #getting the link from the list for next page
    except:
        driver.get(nxt_button[0].get_attribute('href'))

time.sleep(5)
    
# Creating empty list to store data
Reviews = []
Ratings = []

# Scraping data from each url
for i in product_url:
    driver.get(i)
    print("Scraping URL:", i)
    
    try:
        ratings_tag = driver.find_element_by_xpath('//span[@class="user-review-starRating user-review-fourStars"]') #Extracting the ratings
        Ratings.append(ratings_tag.text)
    except NoSuchElementException:
        Ratings.append('No Rating')
    
    try:
        reviews_tag = driver.find_element_by_xpath('//div[@class="user-review-reviewTextWrapper"]') #Extracting the reviews
        Reviews.append(reviews_tag.text.replace('\n',''))
    except NoSuchElementException:
        Reviews.append('No Review')
```

#### Scrapping reviews and ratings from the ecommerce website amazon.in

```python
# Activating the chrome browser
driver=webdriver.Chrome("chromedriver.exe")
driver.maximize_window()
time.sleep(3)

# Opening the homepage of amazon.in
url = "https://www.amazon.in/"
driver.get(url)

time.sleep(2)
# Asking the user to input the keywords he/she wants to search
user_inp = input('Enter the product you want to search : ')

search_bar = driver.find_element_by_xpath('//input[@class="nav-input nav-progressive-attribute"]')    # Finding the search bar using it's xpath
search_bar.clear()  # Clearing the search bar
search_bar.send_keys(user_inp)       # Inputing keyword to search
search_button = driver.find_element_by_xpath('//span[@class="nav-search-submit-text nav-sprite nav-progressive-attribute"]')    # Finding the xpath of search button
search_button.click()      # Clicking the search button
```

```python
# Scraping url of all page
product_url = []
start = 0
end = 19

for page in range(start,end):
    url = driver.find_elements_by_xpath('//a[@class="a-link-normal a-text-normal"]')
    for i in url:
        product_url.append(i.get_attribute('href'))
    nxt_button=driver.find_elements_by_xpath('//li[@class="a-last"]//a')
    try:
        driver.get(nxt_button[1].get_attribute('href'))     #getting the link from the list for next page
    except:
        driver.get(nxt_button[0].get_attribute('href'))
    time.sleep(5)
    
# Creating empty list to store data
Reviews = []
Ratings = []

# Scraping data from each url
for i in product_url:
    driver.get(i)
    print("Scraping URL:", i)
    
    time.sleep(2)
    
    try:
        ratings_tag = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[5]/div[38]/div/div/div[2]/div/div[2]/span[3]/div/div/div[4]/div[3]/div/div/div/div/div[2]/a[1]/i/span')
        Ratings.append(ratings_tag.get_attribute('innerHTML').split(' ')[0])
    except NoSuchElementException:
        Ratings.append('No Rating')
    time.sleep(1)
    try:
        reviews_tag = driver.find_element_by_xpath('//div[@class="a-expander-content reviewText review-text-content a-expander-partial-collapse-content"]//span') #Extracting the reviews
        Reviews.append(reviews_tag.text.replace('\n',''))
    except NoSuchElementException:
        Reviews.append('No Review')
```

```python
print(len(Reviews),len(Ratings))
```

```python
# making dataframe
df_amazon = pd.DataFrame({'Ratings':Ratings, 'Reviews':Reviews})
df_amazon
```

```python
# Saving the csv file

df_amazon.to_csv(r'E:\Flip Robo\Project - 7 - Ratings Prediction\Solution\data\printers_amazon.csv')
```
